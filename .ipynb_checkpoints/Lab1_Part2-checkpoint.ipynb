<p>converting tweets to dataframe format and then looking up information about the users who made those tweets using lookupUsers function </p>

```r
tweet_Frame <- twListToDF(tweets_from_DB)# convert the retrieved tweets list to a R Dataframe
user_Info <- lookupUsers(tweet_Frame$screenName)# gather information about the user(s) who made the  tweets
```

<p>Over here once  again we are converting the user_info list to a dataframe.Of all the users that we have ,we only take into consideration the ones with valid  locations i.e locations which are not null.Then we extract the location from  users with valid locationsa and convert it to a vector .Then we use the table function to count the frequeny of each location in this list.</P>

```r
user_Frame <- twListToDF(user_Info)# convert the list of users to a R Dataframe
users_With_Location <- !is.na(user_Frame$location) # only consider users with location which is not equal to 'NA'
users_with_valid_locations <- twListToDF(user_Info[users_With_Location])# again convert to DF
location_List <- as.vector(users_with_valid_locations$location) # get the locations of users with valid locations
table <- table(location_List) # summarize the location list
top_locations <- table[table>1] # get the locations which has most number of tweets(in this case beyond 2)
top_locations
```

```r
top_locations
```

<p>Here we filter through all the locations and only keep those locations which are within USA and discard the rest.
We also create two new lists one for storing the locations and the other for storing the corresponding number of tweets
from that location.This is necessary as we need to fetch geocodes for each location and table function returns a dataframe with 
two columns namely key and frequency of that key in the list</p><br>
This is how I process each location,we split the location name by a comma,then we check if the string after comma is either 'USA' or 
name of any 52 US states,if it is we include that location in the list of locations else we ignore it

```r
# get the keys (location name) and values (the number of tweets ) from the table 
# Also we are only considering locations which are within USA
# As we are supposed to plot tweet locations on US map we discard all other locations
location_list <- list()
no_tweet_list <- list()
len <- length(top_locations)
counter <- 1
for(name in names(top_locations)){
    if(name!=''){
      if(!is.na(strsplit(name, ",")[[1]][2])){
        country <- strsplit(name, ",")[[1]][2]
        if(country==' USA' || country==' GA' || country==' NV' || country==' CA' || country==' DC' || country==' PA'
          || country==' NY' || country==' IL' || country==' TX' || country==' FL' || country=='IN' || country=='VA'||country=='MA'
          || country==' OR' || country==' WA' || country==' CO' || country==' MI' || country==' ohio' || country==' TN'
          ||country==' NC'||country==' SC'||country==' AL'||country==' MS'||country==' MD'||country==' UT'||country==' OK'
          ||country==' LA'||country==' KY'||country==' KS'||country==' IA'||country==' ID'||country==' HI'||country==' CT'
          ||country==' AR'||country==' AZ'||country==' AK'||country==' NE'||country==' NH'||country==' NM'||country==' OR'
          ||country==' PR'||country==' RI'||country==' SD'||country==' VT'||country==' VI'||country==' WV'||country==' WI' 
          ||country==' NJ'||country==' AK'||country==' AR'||country==' HI'||country==' ID'||country==' KS'||country==' KY' ){
            location_list[[counter]]=name
            no_tweet_list[[counter]]=top_locations[name]
            counter <- counter+1
      }
    }
  }
}
```

<p>Here we get the latitude and longitude pairs for locaiton in location_list using ggmap library and Google maps API</p>

```r
locations <- unlist(location_list, recursive = FALSE, use.names = TRUE)# flatten the list to a vector
tweet_frequncies <- unlist(no_tweet_list, recursive = FALSE, use.names = TRUE)# flatten tge frequency list to a vector
lat_lon<- geocode(locations)# get latitude and longitude for each of the above locations
```

Here we first plot the US map using ggmap library

```r
map<-get_map(location='united states', zoom=4, maptype = "terrain",source='google',color='color')
```

Then we we plot all the latitude longitude pairs using geom_point where the size of the blob is propotional to the number of tweets were made from that particular location


```r
ggmap(map)+geom_point( data=lat_lon, aes(x=lon, y=lat,size = tweet_frequncies), color="green")
```
